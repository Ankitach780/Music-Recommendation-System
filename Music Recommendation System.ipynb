```python
df.drop_duplicates(subset=['Track Name','Artist Name(s)'],inplace=True)
df.shape
```

```python
df=df[['Track Name','Artist Name(s)','Artist Genres','Danceability','Energy','Loudness','Acousticness','Liveness','Valence','Tempo','Time Signature','Instrumentalness']]
df.shape
```

```python
def tolist(string):
    l=[]
    l=string.split(',')
    return l
```

```python
df['Artist Name(s)']=df['Artist Name(s)'].apply(tolist)
```

```python
df['Artist Genres']=df['Artist Genres'].apply(tolist)
```

```python
df.head()
```

```python
df['Artist Name(s)']=df['Artist Name(s)'].apply(lambda x:[i.replace(" ","") for i in x])
df['Artist Genres']=df['Artist Genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df['tags']=df['Artist Name(s)']+df['Artist Genres']
```

```python
df.drop(columns=['Artist Name(s)','Artist Genres'],inplace=True)
```

```python
df['tags']
```

```python
df['tags']=df['tags'].apply(lambda x: " ".join(x))
df['tags']=df['tags'].apply(lambda x: x.lower())
```

```python
df['tags'].dtype
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
vector = vectorizer.fit_transform(df['tags'])
```

```python
vector=vector.toarray()
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity[0]
```

```python
similarity.shape
```

```python
df[df['Track Name']=='I Know You Want Me (Calle Ocho)'].index[0]
```

```python
def recommend(music):
    index=df[df['Track Name']==music].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]][0])
```

```python
recommend("Justified & Ancient - Stand by the Jams")
```
